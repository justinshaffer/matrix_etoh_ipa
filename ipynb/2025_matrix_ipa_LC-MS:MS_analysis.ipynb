```python
# Renaming sample IDs in QZA files from Caitriona
cd /matrix/data/lcms
unzip matrix_lcms_lbm_biom.qza
unzip matrix_lcms_hbm_biom.qza

# Manually renamed and moved BIOM tables to the directory above
# matrix_lcms_lbm_biom.biom
# matrix_lcms_hbm_biom.biom

# Convert original BIOM to TSV
## LBM
biom convert \
  -i /matrix/data/lcms/matrix_lcms_lbm_biom.biom \
  -o /matrix/data/lcms/matrix_lcms_lbm_biom.txt \
  --to-tsv

## HBM
biom convert \
  -i /matrix/data/lcms/matrix_lcms_hbm_biom.biom \
  -o /matrix/data/lcms/matrix_lcms_hbm_biom.txt \
  --to-tsv


# Manually updated sample IDs


# Import TSV files to BIOM
## LBM
biom convert \
  -i /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.txt \
  -o /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.biom \
  --table-type="OTU table" \
  --to-hdf5

## HBM
biom convert \
  -i /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.txt \
  -o /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.biom \
  --table-type="OTU table" \
  --to-hdf5


# Import BIOM to QZA
## LBM
qiime tools import \
  --input-path /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.biom \
  --type 'FeatureTable[Frequency]' \
  --input-format BIOMV210Format \
  --output-path /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.qza

qiime feature-table summarize \
  --i-table /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.qza \
  --o-visualization /matrix/data/lcms/matrix_lcms_lbm_biom_qiita_ids.qzv

# There are 70 samples and 1165 features

## HBM
qiime tools import \
  --input-path /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.biom \
  --type 'FeatureTable[Frequency]' \
  --input-format BIOMV210Format \
  --output-path /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.qza

qiime feature-table summarize \
  --i-table /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.qza \
  --o-visualization /matrix/data/lcms/matrix_lcms_hbm_biom_qiita_ids.qzv

# There are 84 samples and 1165 features
```
